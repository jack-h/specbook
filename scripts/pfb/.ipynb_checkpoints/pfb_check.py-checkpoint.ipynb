````python
def plot_pfb_fir0(n_chan=32, n_tap=4, window="hamming", **kwargs):
    # Generate filter coefficients
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    
    # Compute frequency response
    w, h = freqz(lpf, a=1, worN=8192)
    h_dB = 20 * np.log10 (abs(h))
    
    #plot data
    label = '%i-tap, %s window' % (n_tap, window)
    plt.plot(w/max(w) * n_chan/2, np.abs(h), **kwargs)
    #wmax = np.max(w/max(w) * n_chan/2)
    #rev = h[::-1]
    #plt.plot(w/max(w) * n_chan/2-wmax, np.abs(rev),  **kwargs)

    
def plot_fft_fir0(n_chan=32, n_tap=4, window="hamming", **kwargs):
    # Generate filter coefficients
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    lpf = np.ones_like(lpf) / len(lpf)
    # Compute frequency response
    w, h = freqz(lpf, a=1, worN=8192)
    h_dB = 20 * np.log10 (abs(h))
    plt.plot(w/max(w) * n_chan/2, np.abs(h), **kwargs)
    #wmax = np.max(w/max(w) * n_chan/2)
    #rev = h[::-1]
    #plt.plot(w/max(w) * n_chan/2-wmax, np.abs(rev),  **kwargs)


plt.figure(figsize=(8, 4))
plot_acs(n_chan=128, n_tap=1, window="boxcar", ls='--', c='#333333')
plot_fft_fir0(n_chan=128, n_tap=1, window="boxcar", ls='--', c='#333333')
plot_pfb_fir0(n_chan=128, n_tap=1, window="hann", ls='-', c='#333333')

plt.xlim(0, 4)
plt.ylim(-0.05, 1.05)
plt.ylabel('Amplitude')
plt.xlabel(r'Frequency (Normalized to bin width)')
plt.legend(['FFT', 'Hann-windowed'], frameon=False)
plt.savefig("sinc.pdf")
plt.show()
````

````python
def plot_pfb_fir(n_chan=32, n_tap=4, window="hamming", **kwargs):
    # Generate filter coefficients
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    
    # Compute frequency response
    w, h = freqz(lpf, a=1, worN=8192)
    h_dB = 20 * np.log10 (abs(h))
    
    #plot data
    label = '%i-tap, %s window' % (n_tap, window)
    plt.plot(w/max(w) * n_chan/2, h_dB, **kwargs)
    wmax = np.max(w/max(w) * n_chan/2)
    rev = h_dB[::-1]
    kwargs.pop('label')
    plt.plot(w/max(w) * n_chan/2-wmax, rev,  **kwargs)
    

def plot_acs(n_chan=32, n_tap=4, window="hamming", **kwargs):
    # Generate filter coefficients
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    lpf = np.ones_like(lpf) / len(lpf)
    # Compute frequency response
    w, h = freqz(lpf, a=1, worN=8192)
    h_dB = 10 * np.log10 (abs(h))
    
    #plot data
    label = '%i-tap, %s window' % (n_tap, window)
    plt.plot(w/max(w) * n_chan/2, h_dB, **kwargs)
    wmax = np.max(w/max(w) * n_chan/2)
    rev = h_dB[::-1]
    kwargs.pop('label')
    plt.plot(w/max(w) * n_chan/2-wmax, rev,  **kwargs)

    
def plot_fft_fir(n_chan=32, n_tap=4, window="hamming", **kwargs):
    # Generate filter coefficients
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    lpf = np.ones_like(lpf) / len(lpf)
    # Compute frequency response
    w, h = freqz(lpf, a=1, worN=8192)
    h_dB = 20 * np.log10 (abs(h))
    
    #plot data
    label = '%i-tap, %s window' % (n_tap, window)
    plt.plot(w/max(w) * n_chan/2, h_dB, **kwargs)
    wmax = np.max(w/max(w) * n_chan/2)
    rev = h_dB[::-1]
    kwargs.pop('label')
    plt.plot(w/max(w) * n_chan/2-wmax, rev,  **kwargs)

    
plt.figure(figsize=(8, 5))
plot_acs(n_chan=128, n_tap=1, window="boxcar", 
             label='ACS', ls='dotted', c='#333333')
plot_fft_fir(n_chan=128, n_tap=1, window="boxcar", 
             label='FTF', ls='--', c='#333333')
plot_pfb_fir(n_chan=128, n_tap=8, window="hann", 
             label='8-tap Hann PFB', ls='-', c='#333333')
plt.xlim(-4, 4)
plt.ylim(-100, 3)
#plt.legend(frameon=False)
plt.ylabel('Magnitude (dB)')
plt.xlabel(r'Frequency (Normalized to bin width)')
plt.savefig("fb_comparison.pdf")
plt.show()
````

````python
def plot_impz(n_chan=8, n_tap=4, window="hamming"):
    """ Plot impulse response """
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    l = len(lpf)
    impulse = np.repeat(0.,l); impulse[0] =1.
    x = np.arange(0,l)
    response = lfilter(lpf,1,impulse)
    mm = ['o', 's', 'd', 'p', 'v', '^',  '<', '>']
    cc = ['#000000', '#222222', '#444444', '#666666', '#888888', '#AAAAAA', '#BBBBBB', '#CCCCCC']
    for ii in range(0, 8):
        markerline, stemlines, baseline = plt.stem(x[ii::8], response[ii::8], markerfmt=mm[ii])
        plt.setp(markerline, 'markerfacecolor', '#333333')
        plt.setp(stemlines, 'color', '#333333')
        plt.setp(baseline, 'color', '#000000', 'linewidth', 1)
    plt.ylabel('Amplitude')
    plt.xlabel(r'n (samples)')
    plt.title(r'Impulse response')
````

````python
def plot_impz(n_chan=8, n_tap=4, window="hamming"):
    """ Plot impulse response """
    f_cutoff = 1.0 / n_chan
    lpf = firwin(n_chan * n_tap, cutoff=f_cutoff, window=window)
    l = len(lpf)
    impulse = np.repeat(0.,l); impulse[0] =1.
    x = np.arange(0,l)
    response = lfilter(lpf,1,impulse)
    mm = ['o', 's', 'd', 'p', 'v', '^',  '<', '>']
    cc = ['#000000', '#222222', '#444444', '#666666', '#888888', '#AAAAAA', '#BBBBBB', '#CCCCCC']
    for ii in range(0, 8):
        plt.subplot(8, 1, ii+1)
        markerline, stemlines, baseline = plt.stem(x[:8], response[ii::8], markerfmt=mm[ii])
        plt.setp(markerline, 'markerfacecolor', '#333333')
        plt.setp(stemlines, 'color',  '#333333', 'linewidth', 0)
        plt.setp(baseline, 'color', '#000000', 'linewidth', 0)
        plt.xticks([],[""])
        plt.yticks([], [""])
        #plt.title("Branch %i" % (ii+1))
        plt.ylim(-0.05, 0.2)
        plt.xlim(-0.5, 7.5)
plt.figure(figsize=(4, 6))
plot_impz(n_chan=8, n_tap=8, window="rectangular")
plt.tight_layout()
plt.savefig("fir_filter_decomposition.pdf")
````

### Polyphase filter example

````python
n_chan = 64
n_taps = 4
window = "boxcar"

# Generate noise and window
noise = np.round((np.random.random(n_chan * n_taps) - 0.5) * 256)
#noise = np.sin(np.arange(0, n_chan*n_taps))
window = scipy.signal.get_window(window, n_chan * n_taps)
sinc   = firwin(n_chan * n_taps, cutoff=1.0/n_chan, window="rectangular")
noise_weighted = noise * window * sinc / np.max(window * sinc)

# And plot
plt.figure(figsize=(8, 2))
#plt.subplot(3,1,1)
plt.plot(noise, c='#333333')
plt.xticks([0, 63, 127, 191, 255])
plt.yticks([-128, -64, 0, 64, 127])
plt.xlim(0, n_chan * n_taps)
plt.tight_layout()
plt.savefig("pfb_01.pdf")

plt.figure(figsize=(8, 2))
plt.plot((window * sinc) / np.max(window * sinc), c='#333333')
plt.xticks([0, 63, 127, 191, 255])
plt.xlim(0, n_chan * n_taps)
plt.ylim(-0.5, 1.1)
plt.yticks([-0.5, 0, 0.5, 1])
plt.tight_layout()
plt.savefig("pfb_02.pdf")

plt.figure(figsize=(8, 2))
plt.plot(noise_weighted, c='#333333')
plt.xticks([0, 63, 127, 191, 255])
plt.yticks([-128, -64, 0, 64, 127])
plt.xlim(0, n_chan * n_taps)
plt.axvline(x=63, ls='--', c='#999999')
plt.axvline(x=127, ls='--', c='#999999')
plt.axvline(x=191, ls='--', c='#999999')
plt.tight_layout()
plt.savefig("pfb_03.pdf")


plt.figure(figsize=(8, 2))
tap_data = []
for ii in range(0, n_taps):
    plt.subplot(1, 4, ii+1)
    plt.plot(noise_weighted[ii*n_chan:(ii+1)*n_chan], c='#333333')
    plt.ylim(np.min(noise_weighted), np.max(noise_weighted))

    tap_data.append(noise_weighted[ii*n_chan:(ii+1)*n_chan])
    #plt.xticks([], [])
    if ii == 0:
        plt.yticks([-128, -64, 0, 64, 127])
    else: 
        plt.yticks([-128, -64, 0, 64, 127], ["", "", "", "", ""])
    plt.xticks([0, 63])
    plt.xlim(0, 63)
plt.tight_layout()
plt.savefig("pfb_04.pdf")

plt.figure(figsize=(8, 2))
plt.subplot(1, 4, 1)
plt.plot(np.sum(tap_data, axis=0), c='#333333')
plt.xticks([0, 63])
plt.xlim(0, 63)
plt.ylim(-64, 64)
plt.yticks([-128, -64, 0, 64, 127])
plt.ylim(-160, 160)
plt.tight_layout()
plt.savefig("pfb_05.pdf")
````

````python
n_chan = 128
n_taps = 4
window = "boxcar"
window = scipy.signal.get_window(window, n_chan * n_taps)

n_int = 4096
os_fac = 128
spec_arr = np.zeros((n_int, (n_chan /2) * os_fac + 1))
sinc   = firwin(n_chan * n_taps, cutoff=1.0/n_chan, window="rectangular")
for ii in range(n_int):
    # Generate noise and window
    noise = np.random.random(n_chan * n_taps) - 0.5
    #noise = np.sin(np.arange(0, n_chan*n_taps))
    noise_weighted = noise * window * sinc
    
    noise_summed = noise_weighted.reshape(n_taps, n_chan).mean(axis=0)
    spec = np.abs(np.fft.rfft(noise_summed, len(noise_summed) * os_fac))**2
    spec_arr[ii] = spec
    
spec_avg = spec_arr.mean(axis=0)

plt.plot(spec_avg)

    
````

````python
def db(x): return 10*np.log10(x)

def pfb(data, n_chan, n_tap, window="hamming"):
    
    # Generate filter coefficients
    win_len = n_chan * n_tap
    win_coeffs = scipy.signal.get_window(window, win_len)
    sinc   = firwin(n_chan * n_tap, cutoff=2.0/n_chan, window="rectangular")
    win_coeffs *= sinc
    win_coeffs = win_coeffs.reshape((n_tap, n_chan))
    
    # reshape data for application of filter
    data = data.reshape((data.shape[0] / n_chan, n_chan))
    
    pfb_out = np.zeros((len(data) - n_tap + 1, n_chan / 2 + 1), 
                       dtype='complex128')
    #print pfb_out.shape
    #print data.shape
    
    for ii in range(len(pfb_out)):
        d = data[ii:ii+n_tap]
        d = d * win_coeffs
        #d = d * sinc.reshape((n_tap, n_chan))
        d = np.average(d, axis=0)
        d = np.fft.rfft(d)
        pfb_out[ii] = d
    
    return pfb_out
    
    
n_chan = 256
n_tap = 4
window = "hann"
len_data = n_chan * n_taps * 16

peak_vals = []
for ii in np.arange(0, 400, 0.5):
    data = np.sin(np.linspace(0, 2*np.pi, len_data) * (512+ii))
    d = pfb(data, n_chan, n_tap, window=window)
    d = np.average(np.abs(d**2), axis=0)
    d_max = d[8]
    #print np.argmax(d)
    peak_vals.append(d_max)

plt.plot(db(peak_vals/np.max(peak_vals)))
#for ii in range(len(d)):
#    plt.plot(10*np.log10(np.abs(d[ii])))

plt.ylim(-120, 5)
````

````python
from scipy import signal as ss

def ft_win(window):
    W = np.fft.fftshift(np.fft.fft(window(1024), 32768))
    W_db = 10*np.log10(np.abs(W)**2)
    return W_db - np.max(W_db)

def plt_win(window):
    b = np.linspace(-1, 1, 1024)
    plt.plot(b, window(1024), c='#333333')
    plt.ylim(-0.1, 1.1)
    plt.xticks([-1, 0, 1], ["0", "N/2", "N-1"])
    plt.xlabel("n (samples)")

def plt_ft_win(window):
    W_db = ft_win(window)
    W_mid = W_db[16384-512:16384+512]
    
    b = np.linspace(-16, 16, 1024)
    
    plt.plot(b, W_mid, c='#333333')
    plt.xlim(-16, 16)
    plt.ylim(-120, 5)
    plt.xlabel("Frequency bin")
    plt.xticks([-16, -8, 0, 8, 16])
    

plt.figure(figsize=(10, 4))
plt.subplot(2,4,1)
plt.title("Bartlett")
plt_win(ss.triang)
plt.ylabel("Amplitude")
plt.subplot(2,4,2)
plt.title("Hann")
plt_win(ss.hann)
plt.subplot(2,4,3)
plt.title("Hamming")
plt_win(ss.hamming)
plt.subplot(2,4,4)
plt.title("Blackman-Harris")
plt_win(ss.blackmanharris)
plt.subplot(2,4,5)
plt.ylabel("Magnitude (dB)")
plt_ft_win(ss.triang)
plt.subplot(2,4,6)
plt_ft_win(ss.hann)
plt.subplot(2,4,7)
plt_ft_win(ss.hamming)
plt.subplot(2,4,8)
plt_ft_win(ss.blackmanharris)

plt.tight_layout()
plt.savefig("window_fns.pdf")
````

## Scalloping loss

````python
def db(x): return 10*np.log10(x)
lpf = np.genfromtxt('pfb512coef.txt') / 2**18 / 2**17

xx = np.linspace(0, 6, 3072) - 2

plt.figure(figsize=(8, 4))
d = np.fft.fftshift(np.fft.fft(lpf, 2**20))
l = len(d)
mid = len(d)/2
filter_resp0 = d[mid-2048-1024:mid]**2
mval = np.max(db(filter_resp0))
plt.plot(xx, db(filter_resp0)-mval, ls='dashed', c='#333333')
filter_resp = d[mid-2048:mid+1024]**2
plt.plot(xx, db(filter_resp)-mval, ls='dashed', c='#333333')
filter_resp2 = d[mid-1024:mid+2048]**2
plt.plot(xx, db(filter_resp2)-mval, ls='dashed', c='#333333')
filter_resp3 = d[mid:mid+2048+1024]**2
plt.plot(xx, db(filter_resp3)-mval, ls='dashed', c='#333333')
pp = db(filter_resp + filter_resp2 + filter_resp3 + filter_resp0)
pp -= np.max(pp)
plt.plot(xx,pp, ls='solid', c='#333333', lw=2)
plt.xlim(-2, 2)
plt.ylim(-12, 1)
plt.yticks([-12, -9, -6, -3, 0])
plt.xticks([-2, -1, 0, 1, 2])
plt.xlabel("Frequency (normalized to bin width)")
plt.ylabel("Magnitude (dB)")
plt.savefig("pfb_scalloping.pdf")
plt.show()
````
